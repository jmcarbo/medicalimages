### Exercici 3: contatge i qüantificació

```
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow



im = cv2.imread("/content/medicalimages/img/3 embryos.jpg")
imgray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(imgray, 127, 255, 0)
cv2_imshow(thresh)
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)


cv2.drawContours(im, contours, -1, (0,255,0), 3)
cv2_imshow(im)

for c in contours:
    # Moments
    M = cv2.moments(c)
    # Area of contour
    area = M["m00"]

    if area <= 300.0:
      continue
    # Centroid of contour
    cX = int(M["m10"] / M["m00"])
    cY = int(M["m01"] / M["m00"])

    # Perimeter
    perimeter = cv2.arcLength(c, True)

    print("Àrea: ", f'{area}')
    print("Perímetre: ", f'{perimeter}')

    epsilon = 0.02 * perimeter
    approx = cv2.approxPolyDP(c, epsilon, True)
    print(f'{approx}')

    # Create blank image
    blank_image = np.zeros(imgray.shape, np.uint8)

    # Draw contour in the mask
    cv2.drawContours(blank_image, [approx], -1, (255, 255, 255), -1)

    # Create a mask to select pixels inside the figure
    mask_contour = blank_image == 255

    # Calculate the intensity from the grayscale image
    # filtering out the pixels where in the blank_image their value is not 255
    intensity = np.mean(imgray[mask_contour])
    print(f'{intensity}')

    cv2_imshow(blank_image)

```

### Exercici 4: Gel electroforesis

```
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow



im = cv2.imread("/content/medicalimages/img/4 gel.jpg")
imgray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(imgray, 100, 255, 0)
cv2_imshow(thresh)
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)


cv2.drawContours(im, contours, -1, (0,255,0), 3)
cv2_imshow(im)

for c in contours:
    # Moments
    M = cv2.moments(c)
    # Area of contour
    area = M["m00"]

    if area <= 10.0:
      continue
    # Centroid of contour
    cX = int(M["m10"] / M["m00"])
    cY = int(M["m01"] / M["m00"])

    # Perimeter
    perimeter = cv2.arcLength(c, True)

    print(f'{area}')
    print(f'{perimeter}')

    epsilon = 0.02 * perimeter
    approx = cv2.approxPolyDP(c, epsilon, True)
    print(f'{approx}')

    # Create blank image
    blank_image = np.zeros(imgray.shape, np.uint8)

    # Draw contour in the mask
    cv2.drawContours(blank_image, [approx], -1, (255, 255, 255), -1)

    # Create a mask to select pixels inside the figure
    mask_contour = blank_image == 255

    # Calculate the intensity from the grayscale image
    # filtering out the pixels where in the blank_image their value is not 255
    intensity = np.mean(imgray[mask_contour])
    print(f'{intensity}')

    cv2_imshow(blank_image)
```
